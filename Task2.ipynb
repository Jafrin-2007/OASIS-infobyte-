```python
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
import plotly.express as px

fg = px.bar(
    df,
    x='country name',
    y='2020',
    color='country name',
    title='Unemployment Rate (Country-wise) in 2020',
    template='plotly'
)

fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

```python
top10 = df[['country name', '2020']].sort_values('2020', ascending=False).head(10)

fg = px.bar(
    top10,
    x='country name',
    y='2020',
    color='country name',
    title='Top 10 Countries by Unemployment Rate (2020)',
    template='plotly'
)

fg.show()
```

```python
import plotly.express as px

fig = px.box(
    df,
    x="Region",
    y="Estimated Unemployment Rate (%)",
    color="Region",
    points="all",                 # shows scatter points
    title="Unemployment Rate (State-Wise)",
    template="plotly",
)

fig.update_layout(
    xaxis_title="Region",
    yaxis_title="Estimated Unemployment Rate (%)",
    showlegend=False,              # hides legend (like your image)
    xaxis_tickangle=-45            # rotates state names
)

fig.show()
```

```python
import plotly.express as px

fig = px.scatter(
    df,
    x="Region",
    y="Estimated Unemployment Rate (%)",
    color="Region",
    hover_name="Region",
    facet_col="Region",
    facet_col_wrap=5
)

fig.update_layout(
    title="Unemployment Rate (State/Year) by Scatter Graph",
    xaxis_title="Region",
    yaxis_title="Unemployment Rate (%)",
    legend_title="Region",
    template="plotly_white",
    width=1200,
    height=500
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

# group by state / region and sum (or mean) unemployment rate
state_unemp = (
    df.groupby("Region", as_index=False)["Estimated Unemployment Rate (%)"]
      .sum()
      .sort_values("Estimated Unemployment Rate (%)", ascending=False)
)

fig = px.bar(
    state_unemp,
    x="Region",
    y="Estimated Unemployment Rate (%)",
    color="Region",
    title="Unemployment Rate (State-Wise) by Histogram",
    template="plotly_white"
)

fig.update_layout(
    xaxis_title="Region",
    yaxis_title="Sum of Estimated Unemployment Rate (%)",
    showlegend=True
)

fig.show()
```
